```python
sentences = [
    # Cricket (4)
    "The batter notched up a hundred in the decisive game.",
    "The bowler unleashed a quick yorker.",
    "The side clinched victory by pursuing a massive score.",
    "The skipper opted for a bold field setup.",

    # Cooking (3)
    "The cook whipped up a tasty pasta meal.",
    "Baking demands exact quantities of components.",
    "The formula calls for herbs and crisp veggies.",

    # Cybersecurity (3)
    "Robust passcodes safeguard user profiles.",
    "Encryption shields confidential info from cybercriminals.",
    "Firewalls block illicit entry to networks."
]
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('all-MiniLM-L6-v2')

embeddings = model.encode(sentences)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)
similarity_matrix
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(similarity_matrix, annot=True, cmap='coolwarm')
plt.title("Cosine Similarity Matrix")
plt.show()
```

```python
query = "The bowler took three wickets in one over"

query_embedding = model.encode([query])

similarities = cosine_similarity(query_embedding, embeddings)[0]

# Get top 2 similar sentences
import numpy as np

top_indices = np.argsort(similarities)[-2:][::-1]

for i in top_indices:
    print(sentences[i], "-> Similarity:", similarities[i])
```
